```python
import librosa  # librosa is a python package for audio and music analysis variety of format such as w mp3 mp4 and like that
import torch  # it is machine learning base library used such as computer vision  and nlp  devloped by meta AI

import IPython.display as display
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer  # w2 vectorizer ctc is supported by the notebook on how to fine tune a speech recogination,  Tokenizer is a conversion a text into meanigful  lexical tokens belonging to categories defined
import numpy as np
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
audio, sampling_rate = librosa.load("/content/audio [FDMR.Party].mp3", sr=16000)
```

```python
audio, sampling_rate
```

```python
display.Audio ("/content/audio [FDMR.Party].mp3" , autoplay=True)
```

```python
input_values = tokenizer(audio, return_tensors  = 'pt').input_values
input_values
```

```python
logits = model (input_values).logits #logits which means non-normalized predections
logits
```

```python
predicted_ids = torch.argmax(logits, dim=-1)
```

```python
transcription = tokenizer.decode(predicted_ids[0])
```
